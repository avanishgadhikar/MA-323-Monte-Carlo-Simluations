## Problem 2

```python
#Choice 1
a = 1597
b = 51749
m = 244944
# Choosing seed x_0 = 42
x = 42
import numpy as np
import matplotlib.pyplot as plt
freq = np.zeros(20)
for _ in range(10000):
    x = (a*x + b)%m
    # print(x/m)
    freq[(int)((x/m) // 0.05)] += 1
X_axis = [i*0.05 for i in range(0, 20)]
print(freq)
plt.bar(X_axis, freq, width = 0.02, align = 'edge')
plt.xlabel('Range')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Problem 4

```python
u_i = []
u_i_1 = []
a = 1597
b = 51749
m = 244944
# choosing seed = 42
x = 42
prev = x / m
for cnt in range(17):
    x = (a*x + b)%m
    u_i.append(x/m)
    u_i_1.append(prev)
    prev = x / m


# For N = 1000
N = 1000
for ind in range(17, N):
    temp = u_i[ind - 17] - u_i[ind - 5]
    if(temp < 0):
        temp += 1
    u_i.append(temp)
    u_i_1.append(prev)
    prev = temp;

# Scatter plot of u_i vs. u_i_1
plt.figure(figsize=(10, 6))
plt.scatter(u_i_1, u_i, s=1, color='blue', label='(u_i-1, u_i)')
plt.xlabel('u_i-1')
plt.ylabel('u_i')
plt.title("Scatter Plot of u_i vs. u_i-1")
plt.legend()
plt.grid(True)
plt.show()

# Histogram of u_i
plt.figure(figsize=(10, 6))
plt.hist(u_i, bins=50, edgecolor='black', color='skyblue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of u_i')
plt.grid(True)
plt.show()
    
```
