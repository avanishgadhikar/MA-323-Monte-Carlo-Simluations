# Assignment - 07
Exploring various variance reduction techniques in random variable generation.

```python
# importing dependencies
import numpy as np
import pandas as pd
import math
```

```python
# Defining a Linear Congruence Generator
def uniform_generator(x):
    a = 1597
    c = 2330197
    m = 2**32
    x = (a*x + c)%m
    return x, x/m
```

```python
M = [50, 100, 500, 1000, 5000, 10000, 50000, 100000] # Given values of M

lower = []
upper = []
mean = []
variance = []
ratio = []

lower_old = []
upper_old = []
mean_old = []
variance_old = []

turn = 1

for m in M:
    turn = 1 - turn
    f_U = []
    f_1_U = []
    Z = []
    x1 = 42
    x2 = 37

    for _ in range(m):
        x1, u1 = uniform_generator(x1)
        x2, u2 = uniform_generator(x2)
        f_U.append(math.exp(math.sqrt(u1)))
        f_1_U.append(math.exp(math.sqrt(1 - u2)))
        Z.append((f_U[_] + f_1_U[_])/2)
    
    mean_red = np.mean(Z)
    variance_red = np.var(Z)
    
    mean_U = np.mean(f_U)
    variance_U = np.var(f_U)
    
    mean_1_U = np.mean(f_1_U)
    variance_1_U = np.var(f_1_U)
    
    std_dev = math.sqrt(variance_red)
    lower_bound = mean_red - (1.96*std_dev/(math.sqrt(m)))
    upper_bound = mean_red + (1.96*std_dev/(math.sqrt(m)))
    
    mean.append(mean_red)
    variance.append(variance_red)
    lower.append(lower_bound)
    upper.append(upper_bound)
    width_red = upper_bound - lower_bound
    
    std_dev_old = math.sqrt(variance_U)
    lower_bound_old = mean_U - (1.96*std_dev_old/(math.sqrt(m)))
    upper_bound_old = mean_U + (1.96*std_dev_old/(math.sqrt(m)))
    width_old = upper_bound_old - lower_bound_old
    
    
    
    if(turn == 1):
        mean_old.append(mean_U)
        variance_old.append(variance_U)
        lower_old.append(lower_bound_old)
        upper_old.append(upper_bound_old)
        ratio.append(width_red/width_old) # Comparing ratio new : old
    else:
        mean_old.append(None)
        variance_old.append(None)
        lower_old.append(None)
        upper_old.append(None) 
        ratio.append(None)
    
    print("\nM: ", m)
    print(f"Mean of f(u) = {mean_U}, Variance of f(u) = {variance_U}")
    print(f"Mean of f(1-u) = {mean_1_U}, Variance of f(1-u) = {variance_1_U}")
    print(f"Mean Reduced = {mean_red}, Variance Reduced = {variance_red}")
    # print(f"95% Confidence Interval (old): ({lower_bound_old}, {upper_bound_old})")
    print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
    print("\n--------------------------------------------------------------------")
    
table = {"M": M, "Mean (Lab 07)": mean, "Mean (Lab 06)": mean_old, "Variance (Lab 07)": variance, "Variance (Lab 06)": variance_old, "Lower 95% (Lab 07)": lower, "Lower 95% (Lab 06)": lower_old, "Upper 95% (Lab 07)": upper, "Upper 95% (Lab 06)": upper_old, "Ratio (new:old)": ratio}
```
